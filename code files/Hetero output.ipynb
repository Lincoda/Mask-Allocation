```python
from model_hetero import *
import numpy as np
import pandas as pd
```

### Heterogeneous Groups

```python
vo =0.5
vn=0.7
𝜌L=1
𝜌H=1
σo=0.5
σn=0.7
𝛿o=0.5
𝛿n=0.7
I=0.01
𝛼=0.0138
𝛽H=14*2.4/18
𝛽L=0.1 * 𝛽H
m_vals=np.linspace(0.1,0.8,71)

param=map(lambda m: GRBT_Hetero(m,I_0=I,𝜌L=𝜌L,𝜌H=𝜌H,σo=σo,σn=σn,δo=δo,δn=δn,vo=vo,vn=vn,𝛽L=𝛽L,𝛽H=𝛽H),m_vals)
paramdata= pd.DataFrame.from_records(list(param), columns =['m','p','q','ICphi','ICn',
                                                            'πB','πA','πA0','πB0','πA1','πB1',
                                                            'S3','I3','D_vals','Uphi','Un','Ud'])
paramdata.to_csv("210407 GRBT Hetero.csv")
paramdata
```

```python
result=map(lambda x: SRA1_Hetero(x,I_0=I,σo=σo,σn=σn,δo=δo,δn=δn,𝛽L=𝛽L,𝛽H=𝛽H),m_vals)
result=list(result)
data=pd.DataFrame.from_records(result, columns =['func', 'π1','π2','Uphi','Un','Ud'])
data['m']=m_vals
data=data.rename(columns={'func':'D_vals'})


data.to_csv('210407 SRA1 Hetero.csv',index=False)
data
```

```python
result=map(lambda x: SRA2_Hetero(x,I_0=I,σo=σo,σn=σn,δo=δo,δn=δn,𝛽L=𝛽L,𝛽H=𝛽H),m_vals)
result=list(result)
data=pd.DataFrame.from_records(result, columns =['func', 'π1','π20','π21','Uphi','Un','Ud'])
data['m']=m_vals
data=data.rename(columns={'func':'D_vals'})


data.to_csv('210407 SRA2 Hetero.csv',index=False)
data
```
